## Dataset Overview

> This data set includes information about people who rented bike of ford Gobike system at San-Francisco  .It include only thr recordes of feburary of 2019.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
def chart_name(x ,y ,t):
    plt.title(x, fontsize = 20)
    plt.xlabel(y, fontsize = 15)
    plt.ylabel(t, fontsize = 15)
    plt.show();
    
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Bike.csv')
```

## What is the type of our users?
It is about of 90% of users are Subscriber and 10% are Customer .

```python
# plot the piechart to show the type of user
sorted_counts = df['user_type'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, autopct='%1.0f%%', startangle=90, counterclock = False)
plt.title("The Distribution of Subscription Types", fontsize=20)
plt.axis('square');
```

## What is the gender  of our users?
The majurty of our users are Male .

```python
# plot the piechart to show the type of user
sorted_counts = df['member_gender'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, autopct='%1.0f%%', startangle=90, counterclock = False)
plt.title("The Distribution of Subscription Types", fontsize=20)
plt.axis('square');
```

## What is the average of trip duration ?
The average of trip duration is 12 Minutes.

```python
# Distribution of Age
base_color = sb.color_palette()[0]
plt.figure(figsize = (8,4))

bins = np.arange(0, df['member_age'].max()+5, 5)
sb.histplot(data=df, x='member_age', color=base_color, bins = bins)
plt.xticks(rotation=90)
chart_name('Member age','Count',"Distribution of Age")
```

## What is the Crowded days of Week ?
All weekday are crowded instead of saturday and sunday ,as they are the weekend .

```python
base_color = sb.color_palette()[3]
order = ['Sat', 'Sun','Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sb.countplot(data = df, x = 'end_day', order = order, color = base_color)

chart_name('End_day','Count of Users','Rides on Each End_day')
```

## what is the days that users ride for long time?
Althoug the saturday and sunday have less user,but this days have long trips duration.

```python
# plot the point plot of Start day
sb.pointplot(data=df, x='start_day', y='duration_min', errwidth=0)

chart_name('Start Day' , 'Duration in Miutes' , 'The duration over weak')
```

## Who ride for long time of our users by type?
The Avg of customer is about 25 minutes,but the customer Avg is about 11 minutes per week.

```python
sb.pointplot(data = df, x = 'start_day', y = 'duration_sec', hue = 'user_type', palette = 'Reds', errwidth=0)

chart_name('Days','Mean Trip Dduration in second','Trip duartion by user type in different days')
```

## Who ride for long time of our users by Gender ?
The Avg of Female rides over weak are greater than Male .

```python
# plot the point plot of Duration vs gender


plt.figure(figsize=(10,9))
sb.pointplot(data=df, x='start_day', y='duration_min', hue='member_gender', errwidth=0)

plt.legend(title='Gender')
chart_name('Avg duration ovear weak'  ,'Start Day','Avg Duration in seconed')
```
